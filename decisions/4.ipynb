##### Задача 4 — Проекция признаков на направление

**Вводные данные**:

x = [3, 4]

v = [1, 1]

Задача:

- Реализовать формулу проекции
- Найти проекцию x на v
- Найти ортогональную компоненту

Что найти:

- вектор проекции
- остаток (ошибка аппроксимации)

---
##### **Введение**

##### **Определения**
**Проекция** - это часть вектора, которая дает понять насколько вектор схож с направлением.

**Ортогональная компонента** - часть вектора, выходящая за пределы опущенного перпендикуляра.

##### **Формулы**
$$
proj_w(x)=\frac{x * w}{\|w\|^2}*w
$$

---
##### Переходим к решению

##### Реализация формулы проекции

##### Поиск проекции x на v

```python
x = [3, 4]
y = [1, 1]

def scalar_multiplication(vector1, vector2):
    result = 0
    for index, elem in enumerate(vector1):
        result += elem * vector2[index]
    return result 

def L2(input_data):
    return sum(elem**2 for elem in input_data)

def calculate_projection(x, y):
    result_vector = []
    divide = (scalar_multiplication(x, y) / L2(y))
    
    for elem in y:
        result_vector.append(elem*divide)

    return result_vector

projection = calculate_projection(x, y)

print(f"Проекция: {projection}")
```

##### Ортогональная компонента вектора

```python
print(f"Наша проекция: {projection}")

def orthogonal_projection(vector1, vector2):
    result_vector = []

    for index, elem in enumerate(vector1):
        result_vector.append(elem - vector2[index])

    return result_vector


ort_proj = orthogonal_projection(x, projection)
print(f"Ортогональный вектор: {ort_proj}")
```
