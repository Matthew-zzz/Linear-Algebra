##### **1.3 Косинусная близость**

---
##### Задача 3 — Сравнение объектов по направлению

**Вводные данные**:
Три вектора пользователей:

u1 = [5, 0, 1]

u2 = [4, 1, 0]

u3 = [0, 3, 5]

Задача:

- Реализовать косинусную близость
- Найти, кто ближе всего к u1

Что найти:

- cos(u1, u2), cos(u1, u3)
- вывод о сходстве независимо от масштаба


---
##### **Введение**

##### **Определения**
**Косинусная близость** - мера сходства между двумя векторами, которая вычисляется косинусом угла между ними. Чем меньше угол, тем больше сходство.

Если 1 => маленький угол между векторами(они идентичны)

Если 0 => угол 90°(вектора ортогональны)

Если -1 => большой угол между векторами(они противоположны)
##### **Формулы**

$$
cos(θ) = \frac{A*B}{\|A\|*\|B\|}
$$
A * B - скалярное произведение векторов

||A|| * ||B|| - нормы векторов

---
##### Переходим к решению

```python
from math import sqrt

u1 = [5, 0, 1]
u2 = [4, 1, 0]
u3 = [0, 3, 5]

def scalar_multiplication(vector1, vector2):
    result = 0
    for index, elem in enumerate(vector1):
        result += elem * vector2[index]
    return result 

def L2(input_data):
    return sqrt(sum(elem**2 for elem in input_data))

def cos_similarity(vector1, vector2):
    return scalar_multiplication(vector1, vector2) / (L2(vector1) * L2(vector2))

print(f"Косинусовое сходство u1,u2: {cos_similarity(u1, u2)}")
print(f"Косинусовое сходство u1,u3: {cos_similarity(u1, u3)}")
```

---
##### Вывод по работе
Косинусовая близость используется для нахождения схожести векторов, чтобы понять насколько они близки.

В моем случае:

Векторы u1,u2 - почти идентичны, т.к имеют маленький угол между собой

Векторы u1,u3 - почти ортоганальны, т.к угол около 90°
