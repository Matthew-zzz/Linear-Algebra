##### **1. Базовые операции с векторами**

---
##### Задача 2 — Взвешенная сумма признаков

**Вводные данные**:

Вектор признаков объекта:

x = [2, 3, 5]

Вектор весов модели:

w = [0.4, 0.1, 0.5]

Задача:

- Реализовать скалярное произведение без np.dot
- Посчитать выход линейной модели

Что найти:

- значение скалярного произведения
- вклад каждого признака в итоговый результат
---

##### Уточнения
**Вектор признаков** - вектор характеристик объекта, на которые опираются при обучении модели.

**Вектор весов** - вектор значений, которые показывают важность соответсвующего признака.


---
##### Переходим к решению
##### Реализация скалярного произведения(алгебраического)
**Формула**: 
$$
x = a_1b_1 + a_2b_2 + \dots + a_nb_n = \sum_{i=1}^{n} a_ib_i
$$

```python
x = [2, 3, 5]

w = [0.4, 0.1, 0.5]

def scalar_multiplication(vector1, vector2):
    result = 0
    for index, elem in enumerate(vector1):
        result += elem * vector2[index]
    return result 

print(f"Результат скалярного произведения: {scalar_multiplication(x, w)}")
```

---
##### Вклад каждого признака в итоговый результат

```python
x = [2, 3, 5]

w = [0.4, 0.1, 0.5]

def scalar_multiplication(vector1, vector2):
    result = []
    for index, elem in enumerate(vector1):
        result.append(elem * vector2[index])
    return result 

print(f"Взвешенные признаки: {scalar_multiplication(x, w)}")
```

Показывает что 3 признак имеет наибольшее влияние на результат скалярного произведения
